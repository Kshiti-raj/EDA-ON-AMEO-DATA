# BASIC CLEANING AND CONVERTING OF THE DATA

```python
#converting salary col from float to int: 

df["Salary"]=df['Salary'].astype(int)
```

```python
df['DOJ'] = pd.to_datetime(df['DOJ'])

# get the current date
current_date = dt.date.today()

# update the date column with the current date where date is "present"
df.loc[df['DOL'] == 'present', 'DOL'] = current_date
df['DOL'] = pd.to_datetime(df['DOL'])

# also converting DOB to datetime object
df['DOB'] = pd.to_datetime(df['DOB'])
```

```python
df.head()
```

```python
df = df.drop("Unnamed: 0",axis=1)
```

```python
df['DOB'] = pd.to_datetime(df['DOB'])
current_date= pd.Timestamp(dt.date.today())
df["age_of_candidate"]= (current_date - df["DOB"]).astype('<m8[Y]')
df.head()
```

```python
df["age_of_candidate"]=df['age_of_candidate'].astype(int)
```

```python
df.head()
```

```python
# FINDING OUT HOW MUCH TIME IT TOOK CANDIDATES TO CRACK A JOB AFTER GRADUATION:

df["job_within"] = df["DOJ"].dt.year - df['GraduationYear'] 
df.head()
```

# Q1. Finding out the average salary earned : 

```python
# Filter out rows where Specialization is computer science and engineering :
cs_df = df.loc[df['Specialization'] == 'computer science & engineering']

# Calculate the average salary of Software Engineers, Data Analysts and other designations :
avg_salary = cs_df.loc[cs_df['Designation'].isin(["programmer analyst", "software engineer",
"hardware engineer" ,"associate engineer"])]['Salary'].mean()

# Display the result
print('Average salary of CS graduates in Software Engineering, Data Analysis and Network Engineering is:', avg_salary)
```

# ^ This finding tells us that average salary earned by these people would be 3lac thus proving the hypothesis right.^

```python
from scipy.stats import ttest_ind
```

# Testing difference between gpa of two types of engineer:

```python
# split the data into two groups based on designation
software_gpa = df.loc[df['Designation'] == 'software engineer', 'collegeGPA']
hardware_gpa = df.loc[df['Designation'] == 'hardware engineer', 'collegeGPA']

# conduct a t-test to compare the means
t, p = ttest_ind(software_gpa, hardware_gpa)
print(f"t = {t}, p = {p}")
```

# TO check weather city chosen to work in varies based on gender:

```python
# Grouped data by gender and city of job
grouped_data = df.groupby(['Gender', 'JobCity']).size().unstack()

# heatmap of the grouped data
sns.heatmap(grouped_data, cmap='Blues', annot=True, fmt='g')

#axis labels and a title
plt.xlabel('city of Job')
plt.ylabel('Gender')
plt.title('Distribution of State of Job by Gender')

plt.show()
```
